**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

Seaborn is a popular Python data visualization library that provides a high-level interface for creating informative and attractive statistical graphics. Here are five types of plots that can be created using Seaborn, along with their uses:

* **Scatter plot:** A scatter plot is a plot that shows the relationship between two variables by displaying their values as points on a two-dimensional plot. Seaborn provides scatterplot() function to plot a scatter plot. Scatter plots are useful for visualizing the correlation or association between two continuous variables.

* **Line plot:** A line plot is a plot that displays the change in a variable over time or across different categories. Seaborn provides lineplot() function to plot a line plot. Line plots are useful for visualizing trends and patterns in data.

* **Histogram:** A histogram is a plot that shows the distribution of a single variable by dividing the range of values into bins and counting the number of observations that fall into each bin. Seaborn provides histplot() function to plot a histogram. Histograms are useful for visualizing the shape, center, and spread of a distribution.

* **Box plot:** A box plot is a plot that shows the distribution of a variable by displaying the median, quartiles, and outliers. Seaborn provides boxplot() function to plot a box plot. Box plots are useful for visualizing the spread, skewness, and outliers of a distribution, as well as comparing the distribution of a variable across different categories.

* **Heatmap:** A heatmap is a plot that shows the relationship between two variables by displaying their values as a matrix of colors. Seaborn provides heatmap() function to plot a heatmap. Heatmaps are useful for visualizing the patterns and relationships between variables, especially when one variable is categorical and the other is continuous.

**Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.**

**Note: timepoint, signal, event, and region are columns in the fmri dataset.**


```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
fmri
```

```python
# Draw lineplot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
```

**Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.**

**Note: pclass, age, and fare are columns in the titanic dataset.**


```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
# Boxplot of age vs pclass
sns.boxplot(x="pclass", y="age", data=titanic)
```

```python
# Boxplot of fare vs pclass
sns.boxplot(x="pclass", y="fare", data=titanic)
```

**Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.**

```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
sns.histplot(data=diamonds, x="price",  hue="cut")
```

**Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
flights = sns.load_dataset("flights")
flights
```

```python
flights_wide = flights.pivot("month", "year", "passengers")
sns.heatmap(data=flights_wide, cmap="crest")
```
